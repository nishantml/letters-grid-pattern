```python
def n_letter(size,space='',letter='*'):
    matrix = [[' ']*size for i in range(size)]
    
    for i in range(size):
        matrix[i][0] = letter
        matrix[i][i] = letter
        matrix[i][-1] = letter
        
    for i in matrix:
        print(space.join(i))
        

def o_letter(size,space='',letter='*'):
    matrix = [[' ']*size for i in range(size)]
    
    for i in range(1,size-1):
        matrix[i][0] = letter
        matrix[-1][i] = letter
        matrix[0][i] = letter
        matrix[i][-1] = letter
        
    for i in matrix:
        print(space.join(i))
        
def p_letter(size,space='',letter='*'):
    matrix = [[' ']*size for i in range(size)]
    mid = int(size/2)
    for i in range(size):
        matrix[i][0] = letter
        matrix[0][i] = letter
        matrix[mid][i] = letter
    for i in range(mid+1):
        matrix[i][-1] = letter
        
    for i in matrix:
        print(space.join(i))
        
def q_letter(size,space='',letter='*'):
    matrix = [[' ']*(size) for i in range(size)]
    mid = int(size/2)
    for i in range(size-1):
        matrix[i][0] = letter
        matrix[-2][i] = letter
        matrix[0][i] = letter
        matrix[i][-2] = letter
        matrix[-1][-1] = letter
    matrix[2][2] = letter
        
    for i in matrix:
        print(space.join(i))

def r_letter(size,space='',letter='*'):
    count=size-1
    matrix = [[' ']*(size) for i in range(size)]
    mid = int(size/2)
    for i in range(size):
        matrix[i][0] = letter
        matrix[0][i] = letter
        matrix[mid][i]=letter
        
    for i in range(mid):
        matrix[i][size-1] = letter
        matrix[count][count] = letter
        count=count-1
        
        
    for i in matrix:
        print(space.join(i))
        
def s_letter(size,space='',letter='*'):
    matrix = [[' ']*size for i in range(size)]
    mid = int(size/2)
    for i in range(size):
        matrix[0][i] = letter
        matrix[mid][i] = letter
        matrix[-1][i] = letter
        
    for i in range(mid+1):
        matrix[i][0] = letter
        matrix[2+i][-1] = letter
        
    for i in matrix:
        print(space.join(i))
        
def t_letter(size,space='',letter='*'):
    matrix = [[' ']*size for i in range(size)]
    mid = int(size/2)
    for i in range(size):
        matrix[0][i] = letter
        matrix[i][mid] = letter
    
    for i in matrix:
        print(space.join(i))

def u_letter(size,space='',letter='*'):
    matrix = [[' ']*size for i in range(size)]
    mid = int(size/2)
    for i in range(size):
        matrix[i][-1] = letter
        matrix[i][0] = letter
        matrix[-1][i] = letter
    
    for i in matrix:
        print(space.join(i))
        
def v_letter(size,space='',letter='*'):
    col= size*2
    matrix = [[' ']*col for i in range(size)]
    mid = int(col/2)
    for i in range(size):
        matrix[i][i] = letter

    count=4
    for j in range(size-1):
        count = count-1;
        matrix[count][mid] = '*'
        mid = mid+1
    
    
    for i in matrix:
        print(space.join(i))
        
def w_letter(size,space='',letter='*'):
    col = size*2-1
    mid = int(col/2)
    matrix = [[' ']*col for i in range(size)]
    mid = int(col/2)
    right=mid
    left =mid
    for i in range(size):
        matrix[i][0] = letter
        matrix[i][-1] = letter
        matrix[i][right] = letter
        matrix[i][left] = letter
        right = right+1
        left = left -1
    for i in matrix:
        print(space.join(i))
        
def x_letter(size,space='',letter='*'):
    matrix = [[' ']*size for i in range(size)]
    
    count = size-1
    for j in range(size):
        matrix[j][j] = '*'
        matrix[j][count] = '*'
        count -=1
    
    for i in matrix:
        print(space.join(i))
        
def y_letter(size,space='',letter='*'):
    matrix = [[' ']*size for i in range(size)]
    mid = int(size/2)
    count = size-1
    for j in range(size):
        if(j<=mid):
            matrix[j][j] = '*'
        matrix[j][count] = '*'
        count -=1
    
    for i in matrix:
        print(space.join(i))
        
def z_letter(size,space='',letter='*'):
    matrix = [[' ']*size for i in range(size)]
    mid = int(size/2)
    for i in range(size):
        matrix[0][i] = letter
        matrix[-1][i] = letter
        matrix[i][size-i-1] = letter

    
    for i in matrix:
        print(space.join(i))
        
def a_letter(size,space='',letter='*'):
    col = size*2-1
    mid = int(col/2)
    matrix = [[' ']*col for i in range(size)]
    mid = int(col/2)
    row_mid = int(size/2)
    right=mid
    left =mid
    for i in range(size):
        matrix[i][right] = letter
        matrix[i][left] = letter
        right = right+1
        left = left -1
        
    for i in range(2,col-2):
        matrix[row_mid][i] = letter
        
    for i in matrix:
        print(space.join(i))
```
